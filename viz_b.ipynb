```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('iris')
df.head(10)
```

```python
df.info()
```

```python
df.describe()
```

```python
sns.scatterplot(x = 'sepal_length', y = 'petal_length', data=df, hue = 'species')
```

- ##### The iris flower with petal length > 5 are most likely of type virginica.
- ##### The iris flower with a sepal length < 6 and petal length > 2 is most likely of type setosa.
- ##### The iris flower with petal length between 2.5 and 5 is most likely of type versicolor.


```python
sns.set_style('darkgrid')
sns.pairplot(df, hue = 'species')
```

```python
sns.boxplot(x = 'species', y = 'petal_length', data = df)
```

##### We can seperate flowers using petal lenght while we can't do this using the sepal length as we see.


```python
correlation = df.corr()
correlation
```

```python
sns.heatmap(correlation, linewidths=.5, annot=True, cbar=False)
```

##### The sepal length has a positive relation with petal length and petal width. while sepal width has a negative relation with petal length and petal width.
